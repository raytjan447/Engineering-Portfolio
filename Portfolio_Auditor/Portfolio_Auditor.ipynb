```python
import yfinance as yf
import pandas as pd
import sqlite3

tickers = ['RY.TO', 'TSLA', 'RACE']
print ("Starting Data Ingestion.....")

conn = sqlite3.connect ('portfolio.db')
print ("Connecting to SQL Database....")

for t in tickers:
    print (f"Downloading data for: {t}")
    df = yf.download(t, period="1y")
    df_clean = df[['Close']].copy()
    df_clean.columns = ['Price']
    df_clean.to_sql(t,conn, if_exists='replace')
    print (f"Saved {len(df_clean)} rows to Table: '{t}'")

conn.close()

print ("Success: All data in 'portfolio.db'")
```

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('portfolio.db')
tickers = ['RY.TO', 'TSLA', 'RACE']

print ("Daily Risk & Compliance Report")
print ("------------------------------")

for t in tickers:
    query = f"SELECT * FROM '{t}'"
    df = pd.read_sql(query, conn, index_col='Date')

    if df.empty:
        print (f"[FAIL] ❌ {t}: Data table is empty.")
        continue

    df['Return'] = df['Price'].pct_change()
    latest_return = df['Return'].iloc[-1]
    latest_price = df['Price'].iloc[-1]

    if latest_return < -0.02:
        status = "[ALERT] ⚠️ HIGH VOLATILITY"
    else:
        status = "[PASS] ✅ Stable"
    print(f"{status} {t} | Price: ${latest_price:2f} | Change: {latest_return:.2%}")

conn.close()
```
