```python
import yfinance as yf
import pandas as pd

ticker = "CADIDR=X"
df = yf.download(ticker, period="5y")
print(df.head())

df.to_csv('cad_idr_5y.csv')
```

```python
print (df.tail())
```

```python
clean_data = df[['Close']].copy()
clean_data.columns = ['Price_IDR']
clean_data = clean_data.sort_index(ascending=False)

print("Cleaned Data (Newest First):")
print(clean_data.head())
```

```python
import sqlite3

conn = sqlite3.connect ("currency_data.db")
print("Connected to SQL")
clean_data.to_sql('rates', conn, if_exists='replace')
print("Loaded into the 'rates' table")

clean_data.to_csv('final_cad_data.csv', index=True)
conn.close()
```

```python
conn = sqlite3. connect ("currency_data.db")
sql_query = "Select * FROM rates LIMIT 5"
verification_data = pd.read_sql(sql_query, conn)
print("Reading back from SQL Warehouse:")
print(verification_data)

conn.close()
```
